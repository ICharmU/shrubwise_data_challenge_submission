**Task:** Review the original and revised scripts, along with the shrub lists produced by each. Provide a general comparison of both workflows and outputs, highlighting the main strengths and limitations you observe in the scripts as well as in the resulting shrub lists.

# General Comparison
The base IntELiMon script reads in a .las file and applies a cloth simulation filter to separate ground and non-ground objects. The classifications from this separation are then used to rasterize the original input space. A nearest neighbors algorithm is then applied on the classifications to identify relevant clusters after normalization. These clusters are used to identify stems according to verticality and eigentropy constraints. Relevant leaf area-derived attributes are attached to each cluster. Shrubbery (short vegetation) is estimated based on verticality and normalized z coordinates. Crown (branch) metrics are produced for each cluster. Fine fuels (sticks/leaves/small vegetation pieces) are identified and filtered by the number of hours of fuel they would provide.  
  
The revised script performs the same procedures, but it removes tree branches from clustered trees by looking at height characteristics.  

Both scripts create ground-vegetation classifiers and stems clusters, and calculate cluster-wise fuel hours.

# Strengths 
* General vegetation statistics can be used to more accurately cluster shrubs than vanilla nearest neighbors.  
* The ground is ignored when clustering.
* The revised method seems to retain more points for each cluster which is useful for more accurate mapping.


# Limitations
* Stemming is not continuous. Points are clustered for stemming, but there is not necessarily continuity, even after rasterization.
* The processed 3d space is still very space. There is no interpolation.
* Values are hardcoded. This means any trees that are not fully grown may be classified as shrubs. It would be better if clusters could be produced without manual parameter estimation and instead characteristics of the shrubs were used. This is difficult, though, since it would require significant hand labeling for accurate labels.
