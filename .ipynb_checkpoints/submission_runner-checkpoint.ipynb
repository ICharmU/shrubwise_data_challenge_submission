```python
def make_runner_cfg(base_cfg):
    cfg_local = deepcopy(base_cfg)

    # Make feature pipeline use hybrid storage
    cfg_local.features_runtime.storage.enable_local_store = True
    cfg_local.features_runtime.storage.enable_drive_store = True
    cfg_local.features_runtime.storage.use_hybrid_store = True
    cfg_local.features_runtime.storage.fail_if_drive_missing = True

    # Make labeling pipeline use hybrid storage too, so label masks can be rehydrated
    cfg_local.labeling_runtime.storage.enable_local_store = True
    cfg_local.labeling_runtime.storage.enable_drive_store = True
    cfg_local.labeling_runtime.storage.use_hybrid_store = True
    cfg_local.labeling_runtime.storage.fail_if_drive_missing = False

    return cfg_local

def build_runner_pipelines(base_cfg):
    cfg_local = make_runner_cfg(base_cfg)

    feature_pipe = FeaturePipeline(
        cfg_local,
        ops=build_feature_pipeline_ops(),
        pipeline_config=FeaturePipelineConfig(
            canonical_grid_source=cfg_local.features.canonical_grid_source
        ),
    )
    labeling_pipe = LabelingPipeline(
        cfg_local,
        pipeline_config=LabelingPipelineConfig(),
    )

    print("Feature artifact store :", type(feature_pipe.artifact_store).__name__)
    print("Labeling artifact store:", type(labeling_pipe.artifact_store).__name__)

    if not isinstance(feature_pipe.artifact_store, HybridArtifactStore):
        raise RuntimeError("Feature pipeline did not materialize with HybridArtifactStore.")
    if not isinstance(labeling_pipe.artifact_store, HybridArtifactStore):
        raise RuntimeError("Labeling pipeline did not materialize with HybridArtifactStore.")

    return cfg_local, feature_pipe, labeling_pipe

cfg_local, feature_pipe, labeling_pipe = build_runner_pipelines(cfg)
```

```python
import re

CHUNK_NPZ_CACHE = OrderedDict()
MAX_CHUNK_NPZ_CACHE_ITEMS = 12

CHUNK_PATCH_CACHE = OrderedDict()
MAX_CHUNK_PATCH_CACHE_ITEMS = 8

def _lru_get(store, key):
    if key not in store:
        return None
    value = store.pop(key)
    store[key] = value
    return value

def _lru_put(store, key, value, max_items):
    if key in store:
        store.pop(key)
    store[key] = value
    while len(store) > max_items:
        store.popitem(last=False)

def choose_pixel_layer_names_from_df(reg_df: pd.DataFrame, feature_profile: str = "all_sources") -> list[str]:
    if reg_df.empty:
        return []

    profile_to_sources = {
        "naip": {"naip"},
        "naip_als": {"naip", "als"},
        "naip_als_3dep": {"naip", "als", "3dep"},
        "als_3dep": {"als", "3dep"},
        "all_sources": {"naip", "als", "3dep", "rap"},
    }
    wanted_sources = profile_to_sources[feature_profile]

    work = reg_df.copy()
    if "base_layer_name" not in work.columns:
        work["base_layer_name"] = work["layer_name"].astype(str).str.split("::").str[0]

    if "source_name" not in work.columns:
        raise ValueError("Registry is missing source_name.")

    work = work[work["source_name"].isin(wanted_sources)].copy()
    return sorted(work["base_layer_name"].dropna().astype(str).unique().tolist())

def choose_label_artifact_row(site_id: str, resolution_m: float = 1.0) -> pd.Series | None:
    df = label_artifacts_for_site(site_id)
    if df.empty:
        return None

    if "resolution_m" in df.columns:
        df = df[np.isclose(df["resolution_m"].astype(float), float(resolution_m))].copy()

    if df.empty:
        return None

    def _score_row(row):
        source_version = str(row.get("source_version", "")).lower()
        label_variant = str(row.get("label_variant", "")).lower()
        revised = int(("revised" in source_version) or ("revised" in label_variant))
        has_rel = int(pd.notna(row.get("binary_mask_rel_path", None)))
        return (revised, has_rel)

    order = sorted(range(len(df)), key=lambda i: _score_row(df.iloc[i]), reverse=True)
    return df.iloc[order[0]]

def hydrate_path_from_rel(
    *,
    artifact_store,
    rel_path: str | None,
    local_path: str | Path | None,
) -> Path | None:
    if rel_path is None or (isinstance(rel_path, float) and np.isnan(rel_path)):
        return None

    if local_path is not None:
        local_path = Path(local_path)
        if local_path.exists():
            return local_path
    else:
        if isinstance(artifact_store, HybridArtifactStore):
            local_path = artifact_store.local_store.storage_root / str(rel_path)
        else:
            local_path = Path(rel_path)

    local_path.parent.mkdir(parents=True, exist_ok=True)
    pulled = artifact_store.pull(str(rel_path), local_path=local_path)
    return Path(pulled)

def hydrate_label_artifacts(site_id: str, labeling_pipe, resolution_m: float = 1.0) -> dict:
    row = choose_label_artifact_row(site_id, resolution_m=resolution_m)
    if row is None:
        raise FileNotFoundError(f"No label artifact row found for site={site_id}")

    binary_rel = row.get("binary_mask_rel_path", None)
    conf_rel = row.get("confidence_mask_rel_path", None)

    binary_local = row.get("binary_mask_path", None)
    conf_local = row.get("confidence_mask_path", None)

    binary_path = None
    conf_path = None

    if pd.notna(binary_local) and Path(str(binary_local)).exists():
        binary_path = Path(str(binary_local))
    elif pd.notna(binary_rel):
        binary_path = hydrate_path_from_rel(
            artifact_store=labeling_pipe.artifact_store,
            rel_path=str(binary_rel),
            local_path=str(binary_local) if pd.notna(binary_local) else None,
        )

    if pd.notna(conf_local) and Path(str(conf_local)).exists():
        conf_path = Path(str(conf_local))
    elif pd.notna(conf_rel):
        conf_path = hydrate_path_from_rel(
            artifact_store=labeling_pipe.artifact_store,
            rel_path=str(conf_rel),
            local_path=str(conf_local) if pd.notna(conf_local) else None,
        )

    if binary_path is None or not binary_path.exists():
        raise FileNotFoundError(f"Could not hydrate binary label raster for site={site_id}")

    return {
        "artifact_row": row,
        "binary_path": binary_path,
        "confidence_path": conf_path if (conf_path is not None and conf_path.exists()) else None,
    }

def build_site_plan(site_id: str, *, feature_signature: str, feature_profile: str):
    site_assets = feature_pipe.prepare_site_assets_for_pipeline(site_id, force_refresh=False)
    canonical_grid = feature_pipe.ops.build_canonical_grid_for_site(
        site_id,
        artifact_store=feature_pipe.artifact_store,
        assets=site_assets,
        force_refresh=False,
    )
    chunk_manifest = feature_pipe.ops.build_chunk_manifest(
        canonical_grid,
        artifact_store=feature_pipe.artifact_store,
        force_refresh=False,
    )

    registry = load_or_init_stack_registry(
        site_id,
        artifact_store=feature_pipe.artifact_store,
        config_signature=feature_signature,
    )
    reg_df = stack_registry_frame(registry).copy()
    if reg_df.empty:
        raise ValueError(f"Empty feature registry for site={site_id}, feature_signature={feature_signature}")

    reg_df["base_layer_name"] = reg_df["layer_name"].astype(str).str.split("::").str[0]
    reg_df["chunk_id_parsed"] = reg_df["layer_name"].astype(str).str.split("::").str[-1]

    selected_layers = choose_pixel_layer_names_from_df(reg_df, feature_profile=feature_profile)
    selected_rows = reg_df[reg_df["base_layer_name"].isin(selected_layers)].copy()

    manifest_df = chunk_manifest_frame(chunk_manifest).copy()
    rows = []
    for _, mrow in manifest_df.iterrows():
        chunk_id = str(mrow["chunk_id"])
        chunk_rows = selected_rows[selected_rows["chunk_id_parsed"] == chunk_id].copy()
        missing = len(set(selected_layers) - set(chunk_rows["base_layer_name"].tolist()))
        rows.append({
            "site_id": site_id,
            "chunk_id": chunk_id,
            "row_start": int(mrow["row_start"]),
            "row_end": int(mrow["row_end"]),
            "col_start": int(mrow["col_start"]),
            "col_end": int(mrow["col_end"]),
            "height": int(mrow["row_end"] - mrow["row_start"]),
            "width": int(mrow["col_end"] - mrow["col_start"]),
            "n_rows": int(len(chunk_rows)),
            "n_missing_layers": int(missing),
            "is_chunk_usable": bool(len(chunk_rows) > 0 and missing == 0),
        })
    chunk_availability_df = pd.DataFrame(rows)

    label_bundle = hydrate_label_artifacts(site_id, labeling_pipe)

    return {
        "site_id": site_id,
        "feature_signature": feature_signature,
        "feature_profile": feature_profile,
        "feature_pipe": feature_pipe,
        "labeling_pipe": labeling_pipe,
        "site_assets": site_assets,
        "canonical_grid": canonical_grid,
        "chunk_manifest": chunk_manifest,
        "chunk_manifest_df": manifest_df,
        "registry": registry,
        "reg_df": reg_df,
        "selected_rows": selected_rows,
        "selected_layers": selected_layers,
        "chunk_availability_df": chunk_availability_df,
        "label_bundle": label_bundle,
    }

def get_chunk_record_from_plan(plan: dict, chunk_id: str):
    for rec in plan["chunk_manifest"].records:
        if str(rec.chunk_id) == str(chunk_id):
            return rec
    raise KeyError(f"chunk_id={chunk_id} not found for site={plan['site_id']}")

def load_family_chunk_npz(*, family_row: pd.Series, artifact_store):
    rel_path = family_row.get("rel_path", None)
    local_path = family_row.get("local_path", None)
    cache_key = str(rel_path) if pd.notna(rel_path) else str(local_path)

    cached = _lru_get(CHUNK_NPZ_CACHE, cache_key)
    if cached is not None:
        return cached

    hydrated_path = None
    if pd.notna(local_path) and Path(str(local_path)).exists():
        hydrated_path = Path(str(local_path))
    elif pd.notna(rel_path) and artifact_store.exists(str(rel_path)):
        target_local = Path(str(local_path)) if pd.notna(local_path) else artifact_store.local_store.storage_root / str(rel_path)
        target_local.parent.mkdir(parents=True, exist_ok=True)
        hydrated_path = artifact_store.pull(str(rel_path), local_path=target_local)
    else:
        raise FileNotFoundError(f"Chunk NPZ unavailable. rel_path={rel_path} local_path={local_path}")

    with np.load(hydrated_path, allow_pickle=False) as data:
        payload = {k: np.asarray(data[k]) for k in data.files}

    _lru_put(CHUNK_NPZ_CACHE, cache_key, payload, MAX_CHUNK_NPZ_CACHE_ITEMS)

    try:
        hydrated_path = Path(hydrated_path)
        if hydrated_path.exists() and "_artifact_staging" in str(hydrated_path):
            hydrated_path.unlink()
    except Exception:
        pass

    return payload

def build_chunk_support_mask(label_mask: np.ndarray, confidence: np.ndarray | None = None, dilation_radius_px: int = 24):
    label_pos = label_mask > 0
    if label_pos.any():
        support = morphology.dilation(label_pos, morphology.disk(int(dilation_radius_px)))
    else:
        support = np.zeros_like(label_pos, dtype=bool)

    if confidence is not None:
        support = support | (confidence > 1e-6)

    return support.astype(np.uint8)

def materialize_chunk_feature_label_patch(plan: dict, chunk_id: str, dilation_radius_px: int = 24):
    cache_key = (plan["site_id"], chunk_id, tuple(plan["selected_layers"]), dilation_radius_px)
    cached = _lru_get(CHUNK_PATCH_CACHE, cache_key)
    if cached is not None:
        return cached

    chunk_record = get_chunk_record_from_plan(plan, chunk_id)
    window = chunk_window(chunk_record)
    height = int(chunk_record.height)
    width = int(chunk_record.width)

    chunk_rows = plan["selected_rows"][plan["selected_rows"]["chunk_id_parsed"] == str(chunk_id)].copy()
    if chunk_rows.empty:
        raise ValueError(f"No selected rows for site={plan['site_id']} chunk_id={chunk_id}")

    feature_arrays = {}
    grouped = chunk_rows.groupby(["rel_path", "local_path", "source_name", "family_name"], dropna=False, sort=False)
    for _, sub in grouped:
        family_row = sub.iloc[0]
        npz_payload = load_family_chunk_npz(
            family_row=family_row,
            artifact_store=plan["feature_pipe"].artifact_store,
        )

        for _, row in sub.iterrows():
            layer_name_full = str(row["layer_name"])
            base_layer_name = str(row["base_layer_name"])

            if layer_name_full in npz_payload:
                arr = np.asarray(npz_payload[layer_name_full], dtype=np.float32)
            elif base_layer_name in npz_payload:
                arr = np.asarray(npz_payload[base_layer_name], dtype=np.float32)
            else:
                raise KeyError(f"{base_layer_name} missing from NPZ payload keys={list(npz_payload.keys())[:10]}")

            expected_shape = (height, width)
            if arr.shape != expected_shape:
                raise ValueError(
                    f"Chunk array shape mismatch for site={plan['site_id']}, chunk_id={chunk_id}, "
                    f"layer={base_layer_name}: expected {expected_shape}, got {arr.shape}"
                )
            feature_arrays[base_layer_name] = arr

    layer_names = list(plan["selected_layers"])
    feature_tensor = np.stack([feature_arrays[name] for name in layer_names], axis=0).astype(np.float32)

    label_bundle = plan["label_bundle"]
    dst_transform = window_transform(window, plan["canonical_grid"].transform)
    dst_crs = plan["canonical_grid"].crs

    label_mask = np.zeros((height, width), dtype=np.uint8)
    with rasterio.open(label_bundle["binary_path"]) as src:
        reproject(
            source=rasterio.band(src, 1),
            destination=label_mask,
            src_transform=src.transform,
            src_crs=src.crs,
            src_nodata=src.nodata,
            dst_transform=dst_transform,
            dst_crs=dst_crs,
            dst_nodata=0,
            resampling=Resampling.nearest,
        )

    confidence = np.ones((height, width), dtype=np.float32)
    if label_bundle.get("confidence_path") is not None:
        confidence = np.zeros((height, width), dtype=np.float32)
        with rasterio.open(label_bundle["confidence_path"]) as src:
            reproject(
                source=rasterio.band(src, 1),
                destination=confidence,
                src_transform=src.transform,
                src_crs=src.crs,
                src_nodata=src.nodata,
                dst_transform=dst_transform,
                dst_crs=dst_crs,
                dst_nodata=0.0,
                resampling=Resampling.bilinear,
            )

    support_mask = build_chunk_support_mask(
        label_mask=label_mask,
        confidence=confidence,
        dilation_radius_px=dilation_radius_px,
    )

    out = {
        "site_id": plan["site_id"],
        "chunk_id": str(chunk_id),
        "window": window,
        "dst_transform": dst_transform,
        "layer_names": layer_names,
        "feature_arrays": feature_arrays,
        "feature_tensor": feature_tensor,
        "label_mask": label_mask.astype(np.uint8),
        "confidence": confidence.astype(np.float32),
        "support_mask": support_mask.astype(np.uint8),
        "height": height,
        "width": width,
        "profile": plan["canonical_grid"].profile(dtype="float32", count=1, nodata=np.nan),
    }

    _lru_put(CHUNK_PATCH_CACHE, cache_key, out, MAX_CHUNK_PATCH_CACHE_ITEMS)
    return out
```

# Build Small chunk set

```python
import time
from dataclasses import dataclass
from pathlib import Path
import pandas as pd
from tqdm.auto import tqdm

@dataclass
class SubmissionSitePlan:
    site_id: str
    canonical_grid: any
    chunk_manifest: any
    chunk_df: pd.DataFrame
    label_bundle: dict | None
    selected_chunk_ids: list[str]

def build_submission_site_plan(
    *,
    site_id: str,
    feature_pipe,
    labeling_pipe=None,
    choose_top_k_chunks: int = 4,
    require_labels: bool = True,
):
    t0 = time.perf_counter()

    site_assets = feature_pipe.prepare_site_assets_for_pipeline(site_id, force_refresh=False)
    canonical_grid = feature_pipe.ops.build_canonical_grid_for_site(
        site_id,
        artifact_store=feature_pipe.artifact_store,
        assets=site_assets,
        force_refresh=False,
    )
    chunk_manifest = feature_pipe.ops.build_chunk_manifest(
        canonical_grid,
        artifact_store=feature_pipe.artifact_store,
        force_refresh=False,
    )

    chunk_df = pd.DataFrame([r.__dict__ for r in chunk_manifest.records])

    label_bundle = None
    if require_labels and labeling_pipe is not None:
        label_bundle = read_label_bundle_for_site(
            site_id,
            resolution_m=1.0,
            labeling_pipe=labeling_pipe,
        )

    selected_chunk_ids = chunk_df["chunk_id"].astype(str).tolist()

    # optional label-aware prioritization
    if label_bundle is not None:
        scored = []
        for rec in tqdm(chunk_manifest.records, desc=f"{site_id} chunk scoring", leave=False):
            patch = materialize_chunk_feature_label_patch(
                plan={
                    "site_id": site_id,
                    "feature_signature": None,
                    "feature_profile": "all_sources",
                    "feature_pipe": feature_pipe,
                    "label_bundle": label_bundle,
                    "canonical_grid": canonical_grid,
                    "chunk_manifest": chunk_manifest,
                    "selected_rows": pd.DataFrame(),  # not needed for label-only scoring
                },
                chunk_id=rec.chunk_id,
                force_refresh=True,
                show_progress=False,
            )
            score = int((patch["label_mask"] > 0).sum())
            scored.append((rec.chunk_id, score))

        scored = sorted(scored, key=lambda x: x[1], reverse=True)
        selected_chunk_ids = [cid for cid, s in scored if s > 0][:choose_top_k_chunks]

        if not selected_chunk_ids:
            selected_chunk_ids = chunk_df["chunk_id"].astype(str).tolist()[:choose_top_k_chunks]
    else:
        selected_chunk_ids = chunk_df["chunk_id"].astype(str).tolist()[:choose_top_k_chunks]

    print(
        {
            "site_id": site_id,
            "n_chunks_total": len(chunk_df),
            "selected_chunk_ids": selected_chunk_ids,
            "runtime_sec": round(time.perf_counter() - t0, 2),
        }
    )

    return SubmissionSitePlan(
        site_id=site_id,
        canonical_grid=canonical_grid,
        chunk_manifest=chunk_manifest,
        chunk_df=chunk_df,
        label_bundle=label_bundle,
        selected_chunk_ids=selected_chunk_ids,
    )

TRAIN_SITES = ["calaveras-big-trees", "independence-lake"]
VAL_SITE = "sedgwick"
TEST_SITE = "shaver-lake"

site_ids_needed = TRAIN_SITES + [VAL_SITE, TEST_SITE]

submission_site_plans = {}
for site_id in tqdm(site_ids_needed, desc="Build submission site plans"):
    submission_site_plans[site_id] = build_submission_site_plan(
        site_id=site_id,
        feature_pipe=feature_pipeline,
        labeling_pipe=labeling_pipeline if site_id != TEST_SITE else None,
        choose_top_k_chunks=2 if site_id in TRAIN_SITES else 1,
        require_labels=(site_id != TEST_SITE),
    )
```

```python
def resolve_torch_device(requested: str = "auto") -> str:
    requested = str(requested).lower()
    if requested == "cpu":
        return "cpu"
    if requested == "cuda":
        return "cuda"
    if requested != "auto":
        raise ValueError(f"Unsupported TORCH_DEVICE={requested}")

    if not torch.cuda.is_available():
        return "cpu"

    try:
        _ = torch.zeros(1, device="cuda")
        return "cuda"
    except Exception as e:
        print(f"[device fallback] CUDA exists but current torch build cannot use it; falling back to CPU. err={e}")
        return "cpu"

DEVICE = resolve_torch_device(TORCH_DEVICE)
print("Torch device:", DEVICE)

def iter_selected_train_chunks():
    for site_id in TRAIN_SITES:
        for chunk_id in train_chunk_ids[site_id]:
            yield site_id, chunk_id

def iter_selected_val_chunks():
    for site_id in [VAL_SITE]:
        for chunk_id in val_chunk_ids[site_id]:
            yield site_id, chunk_id

def compute_channel_stats(site_plans, iterator):
    sum_c = None
    sumsq_c = None
    total = 0

    items = list(iterator)
    bar = tqdm(items, desc="channel stats")
    for site_id, chunk_id in bar:
        patch = materialize_chunk_feature_label_patch(
            site_plans[site_id],
            chunk_id=chunk_id,
            dilation_radius_px=SUPPORT_DILATION_RADIUS_PX,
        )
        x = patch["feature_tensor"].astype(np.float64)
        x = np.nan_to_num(x, nan=0.0, posinf=0.0, neginf=0.0)

        c, h, w = x.shape
        flat = x.reshape(c, -1)
        if sum_c is None:
            sum_c = flat.sum(axis=1)
            sumsq_c = (flat ** 2).sum(axis=1)
        else:
            sum_c += flat.sum(axis=1)
            sumsq_c += (flat ** 2).sum(axis=1)
        total += flat.shape[1]

    mean = sum_c / max(total, 1)
    var = np.maximum(sumsq_c / max(total, 1) - mean ** 2, 1e-8)
    std = np.sqrt(var)
    return mean.astype(np.float32), std.astype(np.float32)

def normalize_patch_features(feature_tensor: np.ndarray, mean: np.ndarray, std: np.ndarray) -> np.ndarray:
    x = np.nan_to_num(feature_tensor.astype(np.float32), nan=0.0, posinf=0.0, neginf=0.0)
    return ((x - mean[:, None, None]) / std[:, None, None]).astype(np.float32)

def build_torch_batch_from_patch(patch: dict, mean: np.ndarray, std: np.ndarray, device: str):
    x = normalize_patch_features(patch["feature_tensor"], mean, std)
    X = torch.from_numpy(x[None, ...]).float().to(device)
    y = torch.from_numpy(patch["label_mask"][None, None, ...]).float().to(device)
    conf = torch.from_numpy(patch["confidence"][None, None, ...]).float().to(device)
    sup = torch.from_numpy(patch["support_mask"][None, None, ...]).float().to(device)
    return X, y, conf, sup

def dice_from_binary(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    yt = y_true.astype(np.float32).reshape(-1)
    yp = y_pred.astype(np.float32).reshape(-1)
    denom = float(yt.sum() + yp.sum())
    if denom <= 0:
        return 1.0
    return float(2.0 * (yt * yp).sum() / denom)

def eval_patch_metrics(prob_map: np.ndarray, patch: dict, threshold: float = 0.5) -> dict:
    sup = patch["support_mask"].astype(bool)
    if sup.sum() == 0:
        return {"dice": np.nan, "iou": np.nan, "precision": np.nan, "recall": np.nan}

    y_true = patch["label_mask"][sup].astype(np.uint8)
    y_pred = (prob_map[sup] >= threshold).astype(np.uint8)

    return {
        "dice": dice_from_binary(y_true, y_pred),
        "iou": float(jaccard_score(y_true, y_pred, zero_division=0)),
        "precision": float(precision_score(y_true, y_pred, zero_division=0)),
        "recall": float(recall_score(y_true, y_pred, zero_division=0)),
    }

def predict_prob_map(model, patch: dict, mean: np.ndarray, std: np.ndarray, device: str) -> np.ndarray:
    model.eval()
    with torch.no_grad():
        X, _, _, _ = build_torch_batch_from_patch(patch, mean, std, device)
        logits = model(X)
        if logits.ndim == 4 and logits.shape[1] > 1:
            logits = logits[:, :1, :, :]
        prob = torch.sigmoid(logits).cpu().numpy()[0, 0].astype(np.float32)
        return prob
```

# Train small U-Net

```python
train_items = list(iter_selected_train_chunks())
val_items = list(iter_selected_val_chunks())

if len(train_items) == 0:
    raise RuntimeError("No train chunks selected.")

# infer channels from first train patch
first_patch = materialize_chunk_feature_label_patch(
    site_plans[train_items[0][0]],
    chunk_id=train_items[0][1],
    dilation_radius_px=SUPPORT_DILATION_RADIUS_PX,
)
in_channels = int(first_patch["feature_tensor"].shape[0])

channel_mean, channel_std = compute_channel_stats(site_plans, train_items)

model = CustomUNet(in_channels=in_channels, out_channels=1).to(DEVICE)
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)
criterion = nn.BCEWithLogitsLoss(reduction="none")

history = []

for epoch in range(EPOCHS):
    model.train()
    epoch_losses = []

    bar = tqdm(train_items, desc=f"train epoch {epoch+1}/{EPOCHS}")
    for site_id, chunk_id in bar:
        patch = materialize_chunk_feature_label_patch(
            site_plans[site_id],
            chunk_id=chunk_id,
            dilation_radius_px=SUPPORT_DILATION_RADIUS_PX,
        )
        X, y, conf, sup = build_torch_batch_from_patch(patch, channel_mean, channel_std, DEVICE)

        optimizer.zero_grad()
        logits = model(X)
        if logits.ndim == 4 and logits.shape[1] > 1:
            logits = logits[:, :1, :, :]

        loss_map = criterion(logits, y)
        pixel_weight = sup * conf
        denom = pixel_weight.sum()

        if float(denom.detach().cpu().item()) <= 0.0:
            continue

        loss = (loss_map * pixel_weight).sum() / (denom + 1e-8)
        loss.backward()
        optimizer.step()

        epoch_losses.append(float(loss.item()))
        bar.set_postfix({"loss": f"{float(loss.item()):.4f}", "site": site_id, "chunk": chunk_id})

    # quick validation
    val_rows = []
    for site_id, chunk_id in tqdm(val_items, desc=f"val epoch {epoch+1}/{EPOCHS}", leave=False):
        patch = materialize_chunk_feature_label_patch(
            site_plans[site_id],
            chunk_id=chunk_id,
            dilation_radius_px=SUPPORT_DILATION_RADIUS_PX,
        )
        prob_map = predict_prob_map(model, patch, channel_mean, channel_std, DEVICE)
        m = eval_patch_metrics(prob_map, patch, threshold=0.5)
        m["site_id"] = site_id
        m["chunk_id"] = chunk_id
        val_rows.append(m)

    val_df = pd.DataFrame(val_rows)
    hist_row = {
        "epoch": epoch + 1,
        "train_loss": float(np.mean(epoch_losses)) if epoch_losses else np.nan,
        "val_dice": float(val_df["dice"].mean()) if not val_df.empty else np.nan,
        "val_iou": float(val_df["iou"].mean()) if not val_df.empty else np.nan,
        "val_precision": float(val_df["precision"].mean()) if not val_df.empty else np.nan,
        "val_recall": float(val_df["recall"].mean()) if not val_df.empty else np.nan,
    }
    history.append(hist_row)
    print(hist_row)

history_df = pd.DataFrame(history)
display(history_df)

checkpoint_path = OUTPUT_DIR / "tiny_unet_checkpoint.pt"
torch.save(
    {
        "model_state_dict": model.state_dict(),
        "in_channels": in_channels,
        "channel_mean": channel_mean,
        "channel_std": channel_std,
        "train_sites": TRAIN_SITES,
        "val_site": VAL_SITE,
        "test_site": TEST_SITE,
        "feature_signature": FEATURE_SIGNATURE,
        "feature_profile": FEATURE_PROFILE,
        "train_chunk_ids": train_chunk_ids,
        "val_chunk_ids": val_chunk_ids,
        "epochs": EPOCHS,
        "learning_rate": LEARNING_RATE,
        "support_dilation_radius_px": SUPPORT_DILATION_RADIUS_PX,
        "history": history,
    },
    checkpoint_path,
)
print("Saved checkpoint:", checkpoint_path)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_binary_metrics(labels, predictions, threshold=0.5):
    labels = labels.astype(int).reshape(-1)
    predictions = predictions.astype(float).reshape(-1)
    binary_preds = (predictions >= threshold).astype(int)

    tp = int(np.sum((binary_preds == 1) & (labels == 1)))
    tn = int(np.sum((binary_preds == 0) & (labels == 0)))
    fp = int(np.sum((binary_preds == 1) & (labels == 0)))
    fn = int(np.sum((binary_preds == 0) & (labels == 1)))

    accuracy = (tp + tn) / max(len(labels), 1)
    precision = tp / max(tp + fp, 1)
    recall = tp / max(tp + fn, 1)
    f1 = 2 * precision * recall / max(precision + recall, 1e-8)

    specificity = tn / max(tn + fp, 1)
    npv = tn / max(tn + fn, 1)
    fpr = fp / max(fp + tn, 1)
    fnr = fn / max(fn + tp, 1)

    return {
        "tp": tp, "tn": tn, "fp": fp, "fn": fn,
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "specificity": specificity,
        "npv": npv,
        "fpr": fpr,
        "fnr": fnr,
        "threshold": threshold,
    }

def plot_prediction_diagnostics(
    *,
    predictions,
    labels,
    test_site,
    threshold=0.5,
):
    predictions = np.asarray(predictions).reshape(-1)
    labels = np.asarray(labels).astype(int).reshape(-1)
    binary_preds = (predictions >= threshold).astype(int)
    metrics = compute_binary_metrics(labels, predictions, threshold=threshold)

    tp = metrics["tp"]; tn = metrics["tn"]; fp = metrics["fp"]; fn = metrics["fn"]

    tp_preds = predictions[(binary_preds == 1) & (labels == 1)]
    tn_preds = predictions[(binary_preds == 0) & (labels == 0)]
    fp_preds = predictions[(binary_preds == 1) & (labels == 0)]
    fn_preds = predictions[(binary_preds == 0) & (labels == 1)]

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle(f"Prediction Error Analysis - {test_site.upper()}", fontsize=14, fontweight="bold")

    ax = axes[0, 0]
    for arr, label, color in [
        (tp_preds, f"TP (n={len(tp_preds)})", "#2ecc71"),
        (tn_preds, f"TN (n={len(tn_preds)})", "#3498db"),
        (fp_preds, f"FP (n={len(fp_preds)})", "#e74c3c"),
        (fn_preds, f"FN (n={len(fn_preds)})", "#f39c12"),
    ]:
        if len(arr) > 0:
            ax.hist(arr, bins=40, alpha=0.6, label=label, color=color, edgecolor="black")
    ax.axvline(x=threshold, color="black", linestyle="--", linewidth=2)
    ax.set_xlabel("Prediction Confidence")
    ax.set_ylabel("Frequency")
    ax.set_title("Prediction Confidence by Classification Type")
    ax.legend(fontsize=9)
    ax.grid(axis="y", alpha=0.3)

    ax = axes[0, 1]
    data_to_plot = [tp_preds, tn_preds, fp_preds, fn_preds]
    bp = ax.boxplot(data_to_plot, labels=["TP", "TN", "FP", "FN"], patch_artist=True)
    for patch, color in zip(bp["boxes"], ["#2ecc71", "#3498db", "#e74c3c", "#f39c12"]):
        patch.set_facecolor(color)
    ax.set_ylabel("Prediction Confidence")
    ax.set_title("Prediction Confidence Distribution by Type")
    ax.grid(axis="y", alpha=0.3)

    ax = axes[1, 0]
    categories = ["TP", "TN", "FP", "FN"]
    counts = [tp, tn, fp, fn]
    wedges, texts, autotexts = ax.pie(
        counts,
        labels=categories,
        colors=["#2ecc71", "#3498db", "#e74c3c", "#f39c12"],
        autopct="%1.1f%%",
        startangle=90,
        textprops={"fontsize": 10},
    )
    ax.set_title("Distribution of Predictions")
    for autotext in autotexts:
        autotext.set_color("white")
        autotext.set_fontweight("bold")

    ax = axes[1, 1]
    ax.axis("off")
    table_rows = [
        ["Metric", "Value"],
        ["Accuracy", f"{metrics['accuracy']:.4f}"],
        ["Precision", f"{metrics['precision']:.4f}"],
        ["Recall", f"{metrics['recall']:.4f}"],
        ["F1 Score", f"{metrics['f1']:.4f}"],
        ["Specificity", f"{metrics['specificity']:.4f}"],
        ["NPV", f"{metrics['npv']:.4f}"],
        ["FPR", f"{metrics['fpr']:.4f}"],
        ["FNR", f"{metrics['fnr']:.4f}"],
    ]
    table = ax.table(cellText=table_rows, cellLoc="left", loc="center", colWidths=[0.5, 0.3])
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1, 2)
    ax.set_title("Performance Metrics", pad=20, fontsize=12, fontweight="bold")

    plt.tight_layout()
    plt.show()

    # threshold sweep
    thresholds = np.linspace(0, 1, 101)
    precisions, recalls, f1s = [], [], []
    for thresh in thresholds:
        m = compute_binary_metrics(labels, predictions, threshold=thresh)
        precisions.append(m["precision"])
        recalls.append(m["recall"])
        f1s.append(m["f1"])

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle(f"Model Performance Analysis - {test_site.upper()}", fontsize=14, fontweight="bold")

    ax = axes[0]
    ax.plot(thresholds, precisions, linewidth=2, label="Precision")
    ax.plot(thresholds, recalls, linewidth=2, label="Recall")
    ax.plot(thresholds, f1s, linewidth=2, label="F1 Score")
    ax.axvline(x=threshold, color="gray", linestyle="--", alpha=0.5, label="Current Threshold")
    ax.set_xlabel("Confidence Threshold")
    ax.set_ylabel("Score")
    ax.set_title("Precision, Recall, F1 vs Threshold")
    ax.legend()
    ax.grid(True, alpha=0.3)

    ax = axes[1]
    ax.plot(recalls, precisions, linewidth=2.5, color="#9b59b6", marker="o", markersize=3)
    ax.scatter([metrics["recall"]], [metrics["precision"]], s=200, c="red", marker="*", zorder=5)
    ax.set_xlabel("Recall")
    ax.set_ylabel("Precision")
    ax.set_title("Precision-Recall Curve")
    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])
    ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    return {
        "metrics": metrics,
        "thresholds": thresholds,
        "precisions": precisions,
        "recalls": recalls,
        "f1s": f1s,
        "best_f1_threshold": float(thresholds[int(np.argmax(f1s))]),
        "max_f1": float(np.max(f1s)),
    }
```

```python
val_site = VAL_SITE
val_plan = submission_site_plans[val_site]
val_chunk_id = val_plan.selected_chunk_ids[0]

val_patch = materialize_chunk_feature_label_patch(
    plan=build_chunk_plan_dict_for_submission(val_plan, feature_pipeline, labeling_pipeline),
    chunk_id=val_chunk_id,
    show_progress=False,
)

val_prob = predict_unet_probability_map(model, val_patch["feature_tensor"], device=DEVICE)
val_labels = val_patch["label_mask"]

diag = plot_prediction_diagnostics(
    predictions=val_prob.reshape(-1),
    labels=val_labels.reshape(-1),
    test_site=val_site,
    threshold=0.5,
)
```
